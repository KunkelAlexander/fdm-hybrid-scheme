# The finite difference solution

```python
import numpy as np 
import matplotlib.pyplot as plt

L    = 1
N    = 128
dx   = L/(N - 1)
x    = np.arange(0, N) * dx
dt   = 1e-4
v    = 1
t    = 0
tend = 0.3
r    = (v * dt/dx)**2

def func(x, v, t):
    return np.exp(-(x-0.3 - t * v)**2/0.005)

f  = func(x, v,   0)
fm = func(x, v, -dt)

plt.title(r"$f$")
plt.plot(f)
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.show()


while t < tend:
    fp = 2 * f - fm + r * (np.roll(f, -1) - 2 * f + np.roll(f, 1))
    fm = f 
    f  = fp
    t += dt 

plt.title(f"f at t = {t:3.3f} with RMS error = {np.mean(np.abs(f - func(x, v, t))):3.3e}")
plt.plot(f)
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.show()
```
